```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# c d liu t file CSV
# Thay i ng dn nu file khng nm trong th mc hin ti
data = pd.read_csv('dataset/test.csv')

# Kim tra d liu
print(data[['gross', 'log_gross', 'log_gross_bin']].head())

# Thit lp kiu hin th cho biu 
plt.style.use('default')  # S dng style mc nh ca matplotlib

# To figure vi hai subplot xp dc
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 10), sharex=True)

# V boxplot cho log_gross (biu  trn)
sns.boxplot(x='log_gross_bin', y='log_gross', data=data, ax=ax1)
ax1.set_title('Boxplot of Log Gross Revenue by Bin', fontsize=14, pad=10)
ax1.set_xlabel('')  # n nhn trc X v trc X c chia s vi biu  di
ax1.set_ylabel('Log Gross Revenue', fontsize=12)

# V boxplot cho gross (biu  di)
sns.boxplot(x='log_gross_bin', y='gross', data=data, ax=ax2)
ax2.set_title('Boxplot of Gross Revenue by Bin', fontsize=14, pad=10)
ax2.set_xlabel('Log Gross Bin', fontsize=12)
ax2.set_ylabel('Gross Revenue (USD)', fontsize=12)

# nh dng s trn trc y ca gross  d c (hin th dng triu USD)
ax2.ticklabel_format(style='plain', axis='y')
ax2.get_yaxis().set_major_formatter(
    plt.FuncFormatter(lambda x, loc: f"{x/1e6:.1f}M")
)

# iu chnh layout  khng b chng ln
plt.tight_layout()

# Hin th biu 
plt.show()
```
